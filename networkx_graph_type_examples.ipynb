```python
import networkx as nx
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def draw_all(graph):
    nx.draw_circular(graph)
    plt.title("circular")
    plt.show()
    nx.draw_kamada_kawai(graph)
    plt.title("kamada_kawai")
    plt.show()
    nx.draw_random(graph)
    plt.title("random")
    plt.show()
    nx.draw_spectral(graph)
    plt.title("spectral")
    plt.show()
    nx.draw_spring(graph)
    plt.title("spring")
    plt.show()
    nx.draw_shell(graph)
    plt.title("shell")
    plt.show()
```

```python
draw_all(nx.balanced_tree(2,3))
```

```python
draw_all(nx.barbell_graph(4,4))
```

```python
draw_all(nx.complete_graph(10))
```

```python
draw_all(nx.complete_multipartite_graph(3,3,3))
```

```python
draw_all(nx.circular_ladder_graph(4))
```
