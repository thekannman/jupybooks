```python
from bs4 import BeautifulSoup
import requests
import re

feeds = []

url = "https://feeder.co/knowledge-base/uncategorized/rss-list-popular-and-useful-rss-feeds/"
response = requests.get(url)
soup = BeautifulSoup(response.content, "lxml")
for link in soup.findAll('a', attrs={'href': re.compile("^http://")}):
    feed = feedparser.parse(link["href"])
    feed_dict = {}
    feed_dict["href"] = feed.get("href")
    feed_dict["link"] = link.get("href")
    feed_dict["bozo"] = feed.get("bozo")
    feed_dict["num_entries"] = len(feed["entries"])
    feeds.append(feed_dict.copy())
```

```python
df = pd.DataFrame(feeds)
```

```python
for link in df[df["href"].isna()]["link"]:
    print(link)
```
